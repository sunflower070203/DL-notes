1.RNN介绍
专门处理序列数据的神经网络叫做循环神经网络（Recurrent Neural Network，RNN）。RNN在自然语言处理、语音识别、时间序列预测等领域有广泛的应用。与传统的前馈神经网络不同，RNN具有循环连接，可以将之前的输出作为当前输入的一部分，从而能够捕捉序列中的上下文信息。

2.RNN的结构
RNN的基本结构包括输入层、隐藏层和输出层

RNN的计算过程可以表示为：
h_t = f(W_xh * x_t + W_hh * h_{t-1} + b_h) #当前时刻的隐藏状态
y_t = W_hy * h_t + b_y #当前时刻的输出
其中，h_t表示隐藏层的状态，x_t表示输入，W_xh、W_hh和W_hy分别是输入到隐藏层、隐藏层到隐藏层和隐藏层到输出层的权重矩阵，b_h和b_y是偏置项，f是激活函数。

隐藏状态作用
1. 记忆能力：RNN能够通过隐藏状态来记忆之前的输入信息，从而捕捉序列中的上下文关系。
2. 上下文理解：RNN可以利用隐藏状态来理解当前输入在整个序列中的位置和意义，从而更好地进行预测和生成任务。
3. 连接不同时间步：隐藏状态使得RNN能够连接不同时间步的输入和输出，从而实现对序列数据的处理。

词嵌入层
使用词嵌入层（Embedding Layer）将离散的词语转换为连续的向量表示。

工作流程
1. 输入序列：将文本数据转换为词索引序列。
2. 词嵌入：通过词嵌入层nn.Embedding将词索引转换为词向量。
3. RNN层：将词向量输入到RNN层，捕捉序列中的上下文信息。
4. 输出层：根据任务需求，输出相应的结果，如分类标签或生成文本。

nn.Embedding(num_embeddings, embedding_dim)函数的参数说明：
- num_embeddings：词汇表的大小，即不同词语的数量。
- embedding_dim：词向量的维度，即每个词语被表示为多少维的向量。

RNN=nn.RNN(input_size, hidden_size, num_layers)函数的参数说明：
- input_size：输入特征的维度，即词向量的维度。
- hidden_size：隐藏状态的维度，即RNN层输出的维度。
- num_layers：RNN层的数量，即堆叠的RNN层数。

jieba分词器

jieba.lcut(text)函数的参数说明：
- text：需要分词的中文文本字符串。

循环网络层作用
基于 上一次的隐藏状态 + 本次的输入——> 本次的隐藏状态，本次的输出

公式：
h_t = f(W_xh * x_t + W_hh * h_{t-1} + b_h) #当前时刻的隐藏状态
y_t = W_hy * h_t + b_y #当前时刻的输出，选择概率最大的类别作为最终输出

RNN总结
1. 词嵌入层：
    将词语的索引转化为词向量
2. 循环网络层：
    逐步处理词向量，生成 每个时间步 的隐藏状态。
3. 全连接层
    通过线性变换将隐藏状态映射到输出空间，得到最终的预测结果。

```python
# 导包
import torch
import torch.nn as nn

# 大白话: RNN就像是你的大脑, 在看电影的过程中 记住剧情.

# 1. 创建 循环网络层.
# 参1: 词向量的维度, 就像电影里每一帧画面有128个细节(比如: 颜色, 动作, 表情等), RNN(你的大脑)每次看到的就是这128个细节.
# 参2: 隐藏状态向量维度, 你的大脑能记住的剧情信息量有多大, 比如说: 记住角色关系, 前景提要, 值越大, 能存储的剧情记忆就越多.
# 参3: 隐藏层数量, 默认是1, 你的大脑只有1层来处理它, 如果是1, 说明2层大脑接力理解剧情, 第1层处理, 第2层深加工.
rnn = nn.RNN(input_size=128, hidden_size=256, num_layers=1)

# 2. 定义变量, 表示输入的 x
# 参1: 每个句子的词的个数(长度),  假设电影总共有 5帧画面.
# 参2: 句子的数量,              同时有32个人在看这 5帧画面.
# 参3: 词向量的维度.            每一帧画面的细节是128个.
x = torch.randn(size=(5, 32, 128))

# 3. 定义变量, 记录: 上一时刻的隐藏状态.
# 参1: 隐藏层的层数,   RNN的层数, 和num_layers一样, 1层大脑.
# 参2: 句子的数量,     批量大小, 和上述的 x一样, 32个人...
# 参3: 隐藏状态向量维度,  和上述的 hidden_size一样, 256维.
h0 = torch.randn(size=(1, 32, 256))

# 4. 调用RNN处理, 获取到当前时刻的预测值 和 当前的隐藏状态.
# 参1: x, 本次的输入,  参2: h0, 上一次的隐藏状态.

# 返回值1 output: 每个时间步的输出, 包含了所有时间步的隐藏状态.  每看1帧画面, 大脑能记住的剧情 -> 共5帧, 32个人, 256维.
# 返回值2 h1: 最后1个时间步的隐藏状态.                       看完最后1帧画面后, 大脑里的最新剧情记忆 -> 只关注最后1帧.
output, h1 = rnn(x, h0)
print(f'output: {output.shape}')    # [5, 32, 256]
print(f'h1: {h1.shape}')           # [1, 32, 256]
```
